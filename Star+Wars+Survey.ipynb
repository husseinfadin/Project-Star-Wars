# Part 1
We need to specify an encoding because the data set has some characters that aren't in Python's default utf-8 encoding.

```python
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
bolean_respondent = star_wars["RespondentID"].notnull()
star_wars = star_wars[bolean_respondent]
```

## Cleaning and Mapping Yes/No Columns.
The next six columns represent a single checkbox question. The respondent checked off a series of boxes in response to the question

```python
star_wars[["Have you seen any of the 6 films in the Star Wars franchise?",
           "Do you consider yourself to be a fan of the Star Wars film franchise?"]]
```

# Part 3
> Convert each column above so that it only contains the values True and False

```python
yes_no = {"Yes":True,"No":False}
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)
```

```python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)
```

```python
star_wars[[
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
]]
```

# Part 4

We'll need to convert each of these columns to a Boolean, then rename the column something more intuitive. We can convert the values the same way we did earlier, except that we'll need to include the movie title and NaN in the mapping dictionary.

```python
star_wars.iloc[:,3:9]
```

```python
star_wars.iloc[:,8]
```

```python
starwars_frenshies = {"Star Wars: Episode I  The Phantom Menace":True,
                      "Star Wars: Episode II  Attack of the Clones":True,
                      "Star Wars: Episode III  Revenge of the Sith":True,
                      "Star Wars: Episode IV  A New Hope":True,
                      "Star Wars: Episode VI Return of the Jedi":True,
                      "Star Wars: Episode V The Empire Strikes Back":True,
                      np.nan:False}



for series in star_wars.iloc[:,3:9]:
     star_wars[series]= star_wars[series].map(starwars_frenshies)
```

## Part 8
Exploring the Data BY Segments

```python
male = star_wars[star_wars["Gender"] == "Male"]
female = star_wars[star_wars["Gender"] == "Female"]
```

```python
label = ['ranking_1', 'ranking_2', 'ranking_3', 'ranking_4', 'ranking_5',
       'ranking_6']
men_mean = male[male.columns[9:15]].mean()
women_mean = female[female.columns[9:15]].mean()
index = np.arange(len(label))
width = 0.35
fig, ax = plt.subplots()

ax.bar(index - width/2,men_mean,width,label="Men")
ax.bar(index + width/2,women_mean,width,label="Women",color="r")

ax.set_ylabel("Ranks")
ax.set_title("Avrage Ranked between movies grouped by Gender")
ax.set_xlabel(label)
ax.legend()
```

## Observation.
the avrage ranks for male is littel bit higher then the female in the old movies but for the new movies they are very close .

```python
label_2 = ['seen_1', 'seen_2', 'seen_3', 'seen_4', 'seen_5', 'seen_6']
men_sum = male[male.columns[3:9]].sum()
women_sum = female[female.columns[3:9]].sum()
index = np.arange(len(label_2))
width = 0.25
fig, ax = plt.subplots()

ax.bar(index - width/2,men_sum,width,label="Men")
ax.bar(index + width/2,women_sum,width,label="Women",color="r")

ax.set_ylabel("counts")
ax.set_title("Number of watched Movies between Gender")
ax.set_xlabel(label_2)
ax.set_ylim([0,500])
ax.legend(loc='upper left')
```
